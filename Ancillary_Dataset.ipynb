## Landcover data download and visulization
### Landcover is available at https://www.mrlc.gov/data?f%5B0%5D=year%3A2019 (accessed 10/11/2021)

```python
import xarray as xr
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python

# Each number represents a landcover type, detials refers to meta data comes with the raster file. 
da = xr.open_rasterio("ancillary/Reclass_NLCD_Re1kmMajorit.tif")
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
im =ax.imshow(da.variable.data[0],vmin=0,vmax=20)
divider = make_axes_locatable(plt.gca())
cax = divider.append_axes("right", "5%", pad="3%")
plt.colorbar(im, cax=cax)
```

## soil type data download and visulization
### soil type data is available at https://www.nrcs.usda.gov/wps/portal/nrcs/detail/soils/use/?cid=nrcs142p2_054013

```python
# Each number represents a soil suborder type. Details refer to the meta comes with the data. 
da = xr.open_rasterio("ancillary/so2015v2.tif")
print("Triming to US extent")
da = da.where(da.y<50,drop=True).where(da.y>25,drop=True).where(da.x>-125,drop=True).where(da.x<-65,drop=True)
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
im =ax.imshow(da.variable.data[0],vmin=0,vmax=100)
divider = make_axes_locatable(plt.gca())
cax = divider.append_axes("right", "5%", pad="3%")
plt.colorbar(im, cax=cax)
```

## Global lithology data download and visulization
### Global lithology data is available at https://doi.pangaea.de/10.1594/PANGAEA.788537

```python
da = xr.open_rasterio("ancillary/glim_wgs84_0point5deg.tif")
print("Triming to US extent")
da = da.where(da.y<50,drop=True).where(da.y>25,drop=True).where(da.x>-125,drop=True).where(da.x<-65,drop=True)
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
im =ax.imshow(da.variable.data[0],vmin=1,vmax=13)
divider = make_axes_locatable(plt.gca())
cax = divider.append_axes("right", "5%", pad="3%")
plt.colorbar(im, cax=cax)
```
